This is the code for **Credit-Card fraud detection** using Logistic regression
This involves 3 steps...
    -Data processing
    -Model generation
    -Testing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat=pd.read_csv("fraudTest.csv")
dat #Displaying the data
```

We can infer that the columns/parameters : **Unnamed:0** , **trans_num** , **street** has no significance in fraud detection. So we can delete those columns/parameters for logistic regression

```python
dat.drop(columns=["Unnamed: 0", "trans_num", "street"], inplace= True)
dat
```

Generating train data

```python
#Checking the probability of spam
data_train= dat.head(n = 10000)
data_train.is_fraud.value_counts()
print(data_train.shape)
```

```python
train_dat_processed=pd.get_dummies(data=data_train)
x_train=train_dat_processed.drop(columns="is_fraud",axis=1)
y_train=train_dat_processed["is_fraud"]
```

```python
x_train[1:10]
```

Generating test data

```python
data_test = dat.head(n = 5000)
data_test.is_fraud.value_counts()
```

```python
test_dat_processed=pd.get_dummies(data=data_train)
x_test=test_dat_processed.drop(columns="is_fraud",axis=1)
y_test=test_dat_processed["is_fraud"]
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver='liblinear')
model.fit(x_train,y_train)
```

```python
import sklearn.metrics as metrics
prediction=model.predict(x_test)
model_accuracy=metrics.accuracy_score(y_test,prediction)
print(model_accuracy)
```
